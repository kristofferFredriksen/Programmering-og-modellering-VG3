## 7.1

```python
from pylab import *
from math import *
import numpy as np

Y1 = []
Y2 = []
Y3 = []
X = []

def deriverte (funksjon, x, h):
    fder = (funksjon(x+h) - funksjon(x))/h
    return fder

def f(x):
    return 2*x + 1
def g(x):
    return x**2 - 4*x + 5
def h(x):
    return exp (x)

for x in range (2,5):
    stigning1 = deriverte (f, x, 1E-8)
    stigning2 = deriverte (g, x, 1E-8)
    stigning3 = deriverte (h, x, 1E-8)
    
    X.append (x)
    y1 = f(x)
    Y1.append (y1)
    y2 = g(x)
    Y2.append (y2)
    y3 = h(x)
    Y3.append (y3)
    
    print ("For f(x) er stigningen i x=", x, stigning1)
    print ("For g(x) er stigningen i x=", x, stigning2)
    print ("For h(x) er stigningen i x=", x, stigning3)
    print ("")
    
plot (X,Y1, label = "f'(x)")
plot (X,Y2, label = "g'(x)")
plot (X,Y3, label = "h'(x)")
grid ()
show ()
```

## 7.3

```python
antall_planter = []
endring_planter = []
X2 = []

def N(t):
    return 0.2 * exp (t) + 90

for T in range (11):
    Y = N(T)
    antall_planter.append (Y)
    X2.append (T)
    endring = deriverte (N, T, 1E-8)
    endring_planter.append (endring)
    
plot (X2, antall_planter, Color = "salmon")
plot (X2, endring_planter, Color = "greenyellow")
xlabel ("Tid (år)")
legend (["Antall planter", "Endring i planter per år"])
grid ()
show ()
    
```

```python
temp = []
temp_endring = []

def T(t3):
    return 70*exp(-0.065*t3)

X3 = linspace (0,60,1000)

for i in range (1000):
    tder = deriverte (T, X3[i], 1E-8)
    temp_endring.append (tder)
    temp.append (T(X3[i]))

plot (X3, temp, Color = "mediumspringgreen")
plot (X3, temp_endring, Color = "peachpuff")
legend (["Temperatur i Celsius", "Endring i temperatur"])
xlabel ("Tid (minutter)")
grid ()
show ()

print ("Etter ca.", round (X3[42],2) ,"minutter synker temperaturen med ca.", round (deriverte (T,X3[42], 1E-8), 2), "grader per minutt")
```

## 7.12

```python
data = loadtxt ("Datafiler/posisjon.txt", skiprows = 1)
tid = data[:,0]
posisjon = data[:,1]

plot (tid, posisjon, Color = "deepskyblue")
xlabel ("Tid (Sekunder)")
ylabel ("Posisjon (meter)")
legend(["Posisjon (m)"])
grid ()
show ()
```

```python
fart = []

for e in range (len(tid) - 1):
    dyF = posisjon [e+1] - posisjon [e]
    dxF = tid [e+1] - tid [e]
    derivertF = dyF/dxF
    fart.append (derivertF)
    
plot (tid[:-1], fart, Color = "Hotpink")
xlabel ("Tid (sekunder)")
ylabel ("Fart (m/s)")
legend(["Fart (m/s)"])
grid ()
show ()
```

```python
akselerasjon = []

for k in range (len(tid)-2):
    dyA = fart [k+1] - fart [k]
    dxA = tid [k+1] - tid [k]
    derivertA = dyA/dxA
    akselerasjon.append (derivertA)
    
plot (tid[:-2], akselerasjon, Color = "cyan")
xlabel ("Tid (sekunder)")
ylabel ("Akselerasjon (m/sˆ2)")
legend(["Akselerasjon (m/sˆ2)"])
grid ()
show ()

# Her er grafen veldig nøyaktig, men ser veldig rar ut fordi den tar i bruk så mange desimaler
```

## Integrasjon

## 7.14

```python
def s(x):
    return 2*x + 1
def z(x):
    return x**2 - 4*x + 5
def p(x):
    return exp (x)

def rektangel (fr, a, b, nr): # g = funksjon, a = startpounkt, b = sluttpunkt, n = antall rektangler
    A  = 0 # Setter starverdi av areal som 0
    h = (b-a)/nr # beregner bredden til rektanglene (delta x)
    for i in range (nr):
        A = A + fr(a + i*h)
    areal = A * h
    return areal  

print ("Arealet under grafen s =", rektangel (s, 2, 4, 1000))

print ("Arealet under grafen z =", rektangel (z, 2, 4, 10000))

print ("Arealet under grafen p =", rektangel (p, 2, 4, 10000))
```
