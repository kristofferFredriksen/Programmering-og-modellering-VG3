```python
from pylab import *
#f = input ("funksjonen du ønsker deriver: ")
Y = []
X = []
def f(x1):
    return 2*x1+1

def deriverte (f, x1, h):
    fder = (f(x1+h) - f(x1))/h
    return fder

for x in range (2,5):
    stigning = deriverte (f, x, 1E-8)
    y = f(x)
    X.append (x)
    Y.append (Y)
    print ("for x =", x, "er stigningstallet,", stigning,)
    
plot (x,y)
show()
```

```python
from pylab import *
from math import *

antall_planter = []
endring_planter = []
X2 = []

def N(t):
    return 0.2 * exp (t) + 90

for T in range (11):
    Y = N(T)
    antall_planter.append (Y)
    X2.append (T)
    endring = deriverte (N, T, 1E-8)
    endring_planter.append (endring)
    
plot (X2, antall_planter)
plot (X2, endring_planter)
xlabel ("Tid (år)")
legend (["Antall planter", "Endring i planter per år"])
grid ()
show ()
    
```

```python
from pylab import *
from math import *

temp = []
temp_endring = []

def T(t3):
    return 70*exp(-0.065*t3)

X3 = linspace (0,60,1000)

for i in range (1000):
    tder = deriverte (T, X3[i], 1E-8)
    temp_endring.append (tder)
    temp.append (T(X3[i]))
    
plot (X3, temp)
plot (X3, temp_endring)
legend (["Temperatur i Celsius", "Endring i temperatur"])
xlabel ("Tid (minutter)")
grid ()
show ()

print ("Etter ca.", round (X3[42],2) ,"minutter synker temperaturen med ca.", round (deriverte (T,X3[42], 1E-8), 2), "grader per minutt")
```
